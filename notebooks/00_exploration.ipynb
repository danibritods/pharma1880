```python
def parse_notion_property(prop):
    """
    Parses a single Notion property object based on its type.
    """
    prop_type = prop.get('type')

    if prop_type == 'title':
        return prop['title'][0]['plain_text'] if prop['title'] else None
    
    elif prop_type == 'rich_text':
        return ''.join([item['plain_text'] for item in prop['rich_text']])
        
    elif prop_type == 'multi_select':
        return [option['name'] for option in prop['multi_select']]
        
    elif prop_type == 'select':
        return prop['select']['name'] if prop['select'] else None

    elif prop_type == 'status':
        return prop['status']['name'] if prop['status'] else None
        
    elif prop_type == 'number':
        return prop['number']
        
    elif prop_type == 'url':
        return prop['url']

    elif prop_type == 'relation':
        return [item['id'] for item in prop['relation']]
        
    else:
        return "Unsupported property type: " + str(prop_type)

parsed_data = {}

for prop_name, prop_value in row_properties.items():
    parsed_data[prop_name] = parse_property(prop_value)

import json
print(json.dumps(parsed_data, indent=2, ensure_ascii=False))
```
